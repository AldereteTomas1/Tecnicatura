# **2. Tipos de Datos y su Clasificación**
La identificación y clasificación de los datos es un paso clave en minería de datos, ya que determina qué técnicas de limpieza y transformación pueden aplicarse.

2.1. Tipos de Datos según su Estructura
Los datos pueden organizarse de diferentes maneras según su formato y origen.

Estructurados: Datos organizados en filas y columnas. Se almacenan en bases de datos relacionales.   Tabla de clientes con nombres, edades y saldos bancarios.

No estructurados:    Datos sin formato predefinido, difíciles de almacenar en bases SQL.                   Imágenes, videos, audios, correos electrónicos.

Semiestructurados:   Datos que tienen etiquetas o metadatos que facilitan su organización.                 Archivos JSON, XML, documentos HTML.

Ejemplo en Python – Cargando diferentes tipos de datos

```python
import pandas as pd

# Cargar datos estructurados (desde CSV)
df_csv = pd.read_csv("clientes.csv")
# Cargar datos semiestructurados (JSON)
df_json = pd.read_json("clientes.json")
# Inspeccionar los datos

print(df_csv.head())
print(df_json.head())
```

# **2.2. Tipos de Datos según su Naturaleza**
Los datos también pueden clasificarse según su naturaleza matemática y uso en modelos analíticos.

Numéricos:     Valores que representan cantidades medibles. Precio, peso, temperatura.

Categóricos:    Representan etiquetas o nombres de categorías.                                 Género (Masculino/Femenino), Estado Civil.

Ordinales:   Categorías con un orden lógico, pero sin una distancia uniforme entre ellas.   Nivel educativo (Primaria, Secundaria, Universitaria).

Conclusión:

• Los datos numéricos pueden usarse directamente en modelos de predicción.

• Los datos categóricos deben transformarse antes de usarse en algoritmos de Machine Learning.


```python
import pandas as pd


# Crear un DataFrame con distintos tipos de datos
df = pd.DataFrame({
"Edad": [25, 30, 35, 40, 45],
"Ingreso": [50000, 60000, 55000, 75000, 90000],
"Estado_Civil": ["Soltero", "Casado", "Viudo", "Casado", "Soltero"],
"Nivel_Educativo": ["Secundaria", "Universitaria", "Secundaria", "Primaria", "Universitaria"]
})

# Inspeccionar los tipos de datos
print(df.dtypes)
```

#Transformación de Datos Categóricos

Algunos algoritmos no pueden trabajar con datos categóricos, por lo que es necesario convertirlos en datos numéricos.

**Codificación Ordinal**

Usada cuando las categorías tienen un orden específico.

Ejemplo - Convertir Nivel Educativo en Números

```python
from sklearn.preprocessing import OrdinalEncoder

# Definir orden de categorías

orden_educativo = [["Primaria", "Secundaria", "Universitaria"]]

# Aplicar codificación ordinal
encoder = OrdinalEncoder(categories=orden_educativo)
df["Nivel_Educativo_Cod"] = encoder.fit_transform(df[["Nivel_Educativo"]])
print(df[["Nivel_Educativo", "Nivel_Educativo_Cod"]])
```

2.3.2. Codificación One-Hot Encoding

Se usa cuando las categorías no tienen un orden jerárquico.

Ejemplo – Convertir Estado Civil en Variables Binarias

```python
from sklearn.preprocessing import OneHotEncoder

# Aplicar One-Hot Encoding
encoder = OneHotEncoder(sparse=False, drop="first")
encoded_data = encoder.fit_transform(df[["Estado_Civil"]])

# Convertir a DataFrame
df_encoded = pd.DataFrame(encoded_data, columns=encoder.get_feature_names_out(["Estado_Civil"]))
df = pd.concat([df, df_encoded], axis=1)
print(df.head())
```

Ejercicio para la Sección 2
Objetivo: Aplicar la clasificación y transformación de datos categóricos.

Ejercicio 1: Clasificación y Transformación de Datos
Supongamos que tenemos el siguiente conjunto de datos:

```python
import pandas as pd

from sklearn.preprocessing import OrdinalEncoder, OneHotEncoder

# Crear DataFrame

data = {
"ID": [1, 2, 3, 4],
"Nombre": ["Juan", "María", "Pedro", "Ana"],
"Edad": [32, 28, 40, 35],
"Categoria_Cliente": ["Oro", "Plata", "Platino", "Oro"],

"Estado_Civil": ["Soltero", "Casado", "Viudo", "Casado"]
}
df = pd.DataFrame(data)
# Codificación ordinal para "Categoría Cliente"

orden_categoria = [["Plata", "Oro", "Platino"]]
encoder_ordinal = OrdinalEncoder(categories=orden_categoria)
df["Categoria_Cod"] = encoder_ordinal.fit_transform(df[["Categoria_Cliente"]])
# One-Hot Encoding para "Estado Civil"
encoder_onehot = OneHotEncoder(sparse_output=False, drop="first") # Se cambia 'sparse' a 'sparse_output'
encoded_estado = encoder_onehot.fit_transform(df[["Estado_Civil"]])
df_encoded = pd.DataFrame(encoded_estado, columns=encoder_onehot.get_feature_names_out(["Estado_Civil"]))
df = pd.concat([df, df_encoded], axis=1)
print(df.head())
```

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder

# Datos originales
df = pd.DataFrame({
"Cliente": ["A", "B", "C", "D"],
"Estado_Cuenta": ["Pagado", "Moroso", "Pagado", "Moroso"]
})

# Convertir texto a números
encoder = LabelEncoder()
df["Estado_Cuenta_Num"] = encoder.fit_transform(df["Estado_Cuenta"])
print(df)
```
