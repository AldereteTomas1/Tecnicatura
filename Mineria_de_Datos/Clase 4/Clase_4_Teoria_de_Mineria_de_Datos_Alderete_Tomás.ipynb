Ejemplo en Python – Diferentes Tipos de Visualización



```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Crear datos
df = pd.DataFrame({
    "Edad":      [25, 32, 40, 35, 50, 45, 29, 37, 41, 60],
    "Ingresos":  [30000, 45000, 60000, 55000, 80000, 75000, 40000, 58000, 72000, 90000],
    "Categoría": ["Joven", "Adulto", "Adulto", "Adulto", "Senior", "Senior", "Joven", "Adulto", "Adulto", "Senior"]
})

# Histograma de edades
sns.histplot(df["Edad"], bins=5, kde=True)
plt.title("Distribución de Edades")
plt.show()

# Boxplot de ingresos
sns.boxplot(x=df["Ingresos"])
plt.title("Boxplot de Ingresos")
plt.show()

# Scatter Plot entre Edad e Ingresos
sns.scatterplot(x=df["Edad"], y=df["Ingresos"], hue=df["Categoría"])
plt.title("Relación entre Edad e Ingresos")
plt.show()
```

Conclusión: Seaborn es ideal para gráficos estadísticos más complejos, como histogramas con KDE, boxplots y pairplots.

Ejemplo en Python – Pairplot con Seaborn

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Cargar dataset de pingüinos de Seaborn
df = sns.load_dataset("penguins")

# Crear Pairplot
sns.pairplot(df, hue="species", diag_kind="kde")
plt.show()
```

Ejemplo en Python – Heatmap de correlaciones



```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Cargar dataset
df = sns.load_dataset("penguins").dropna()

# Calcular matriz de correlación
correlaciones = df.corr(numeric_only=True)

# Crear Heatmap
sns.heatmap(correlaciones, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Mapa de Correlación entre Variables")
plt.show()
```

Ejemplo en Python – Violin Plot

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Cargar dataset
df = sns.load_dataset("penguins")

# Crear Violin Plot
sns.violinplot(x="species", y="bill_length_mm", data=df, palette="muted")
plt.title("Distribución del Largo del Pico por Especie")
plt.show()
```

Instrucciones:

Carga el dataset de diamantes de Seaborn (diamonds).

Crea un Pairplot para analizar la relación entre precio, quilates y profundidad.

Genera un Heatmap para visualizar la correlación entre las variables numéricas.

Crea un Violin Plot para comparar la distribución del precio en cada categoría de corte (cut).

Escribe una conclusión basada en los gráficos

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Cargar dataset
df = sns.load_dataset("diamonds")

# 1. Pairplot entre precio, quilates y profundidad
sns.pairplot(df[["price", "carat", "depth"]])
plt.show()

# 2. Heatmap de correlaciones
correlaciones = df.corr(numeric_only=True)
sns.heatmap(correlaciones, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Mapa de Correlación de Diamantes")
plt.show()

# 3. Violin Plot para comparar precio según corte
sns.violinplot(x="cut", y="price", data=df, palette="muted")
plt.title("Distribución del Precio por Tipo de Corte")
plt.show()
```

¿Qué variables están más correlacionadas con el precio?

Observando el Mapa de correlación de Diamantes, podemos observar que las variables: carat "quilates" y las dimensiones de los diamantes "x, y, z" están altamente corelacionadas positivamente con el precio, es decir que cuando aumenta alguna de las variables mencionadas, aumenta la otra.


¿Cómo varía la distribución del precio en los diferentes tipos de corte?

Podemos ver que en el gráfico de Boxplot Multiple hay una gran concentración de los datos en todos los casos, pero podemos observar que el corte "Ideal" hay una concentración en precios más bajos a comparación de las demás variables que tienen una disperción un poco mayor, sin embargo vemos como todas las variables en algún registro consiguen alcanzar los valores más altos llegando a casi 20.000, en este sentido ninguna variable destaca.

¿Qué otros insights puedes extraer de estos gráficos?

Podemos observar la linea blanca en los Boxplot Multiples lo que nos indica la media y entre los cortes vemos que el corte "Fair" presenta una mayor media en comparación al resto, es decir tiende a tener un precio más elevado que el resto.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Crear DataFrame
df = pd.DataFrame({
    "Cliente":   [1, 2, 3, 4, 5, 6, 7, 8],
    "Edad":      [25, 40, 35, 50, 30, 45, 38, 29],
    "Ingresos":  [50000, 60000, 70000, 80000, 45000, 85000, 72000, 48000],
    "Compras":   [5, 15, 10, 20, 7, 18, 12, 6],
    "Categoría": ["A", "B", "C", "A", "B", "C", "A", "B"]
})

# 1) Análisis Exploratorio
sns.histplot(df["Edad"], bins=5, kde=True)
plt.title("Distribución de Edades")
plt.show()

sns.boxplot(x=df["Compras"])
plt.title("Boxplot de Compras")
plt.show()

sns.scatterplot(x=df["Ingresos"], y=df["Compras"])
plt.title("Relación entre Ingresos y Compras")
plt.show()

# 2) Visualización Avanzada
sns.pairplot(df[["Edad", "Ingresos", "Compras"]])
plt.show()

# Mapa de calor de correlaciones
correlaciones = df.corr(numeric_only=True)
sns.heatmap(correlaciones, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Mapa de Correlación de Variables")
plt.show()

# Violin Plot de Compras por Categoría
sns.violinplot(x="Categoría", y="Compras", data=df, palette="muted")
plt.title("Distribución de Compras por Categoría")
plt.show()
```
