```python
#Ejercicio 5: Crea una matriz de 4x4 con valores aleatorios entre 0 y 1.

#np.random.default_rng(): numeros aleatorios
#rng.random((4,4)): matriz de dimensiones x, y
#por default vienen valores entre 0 y 1

rng = np.random.default_rng()
a = rng.random((4,4))
a
```

```python
#Ejercicio 7: Multiplica dos matrices 3x3 de enteros aleatorios entre 1 y 10.

#np.random.randint(): genera numeros aleatorios entre 1 y 11 (sin tomar el 11), en una matriz de 3x3
a = np.random.randint(1, 11, size=(3, 3))
b = np.random.randint(1, 11, size=(3, 3))


print(a)
print("\n") #Bajar un renglon
print(b)
print("\n") #Bajar un renglon
print(a * b)
```

```python
#Ejercicio 8: Crea un array de 15 valores aleatorios entre 1 y 100 e imprime su máximo, mínimo y promedio.

a = np.random.randint(1, 101, size=15)
print(a)
print("\n") #Bajar un renglon
print("El maximo es: ", a.max()) #a.max valor maximo
print("El mínimo es: ", a.min()) #a.min valor minimo
print("El promedio es de: ", a.mean()) #a.mean valor promedio
```

```python
#Ejercicio 9: Genera una matriz 5x5 con valores del 1 al 25 y extrae su diagonal.

# .reshape hace que la matriz de 25 valores se distribuyan entre los espacios de la matriz de 5 x 5
x = np.arange(1, 26).reshape(5, 5)

# np.diag extrae los valores de la diagonal de la matriz "x"
diagonal = np.diag(x)
print(x)

print("\n") #Bajar un renglon
print(diagonal)
```

```python
#Ejercicio 10: Dada una matriz 5x5 de enteros aleatorios entre 1 y 20, calcula la suma de cada fila y columna.

a = np.random.randint(1, 21, size=(5, 5))
print(a)
print("\n") #Bajar un renglon
print(a.sum(axis=0))
print("\n") #Bajar un renglon
print(a.sum(axis=1))
```
